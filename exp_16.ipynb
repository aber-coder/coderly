```python
random_floats = np.random.rand(100)
plt.hist(x=random_floats,bins=10,color="blue")
plt.show()
```

```python
data1 = np.random.randint(1, 101, size=100)
data2 = np.random.randint(1, 101, size=100)
data3 = np.random.randint(1, 101, size=100)


plt.boxplot([data1, data2, data3])
plt.title("Box Plot")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']  
values = [10, 15, 7, 20]
plt.title("Bar Plot")
plt.bar(categories, values, color='lightblue')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(1, 100, size=len(labels))
values 
plt.pie(values, labels=labels)
plt.title("Pie Chart")
plt.show()
```
