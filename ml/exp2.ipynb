```python
import pandas as pd
import numpy as np

# Load the dataset from the CSV file
df = pd.read_csv('weather.csv')

# Separate positive and negative examples based on the target column 'enjoysport'
positive_examples = df[df['enjoysport'] == 'yes'].drop(columns='enjoysport')
negative_examples = df[df['enjoysport'] == 'no'].drop(columns='enjoysport')

# Initialize the most specific and most general hypotheses
specific_hypothesis = positive_examples.iloc[0].tolist()  # Start with the first positive example
general_hypothesis = [['*'] * len(specific_hypothesis)]  # General hypothesis: '*' for all attributes

# Helper functions for generalization and specialization
def generalize(hypothesis, example):
    return [h if h == e else '?' for h, e in zip(hypothesis, example)]

def specialize(hypothesis, example):
    return [e if h == '?' else h for h, e in zip(hypothesis, example)]

# Process positive examples to adjust hypotheses
for example in positive_examples.values:
    general_hypothesis = [g for g in general_hypothesis if all(g[i] == '?' or g[i] == example[i] for i in range(len(g)))]
    specific_hypothesis = [generalize(specific_hypothesis, example)]

# Process negative examples to adjust hypotheses
for example in negative_examples.values:
    general_hypothesis = [g for g in general_hypothesis if not all(g[i] == '?' or g[i] == example[i] for i in range(len(g)))]
    specific_hypothesis = [specialize(specific_hypothesis[0], example)]

# Output final hypotheses
print("Final Specific Hypothesis:", specific_hypothesis)
print("Final General Hypothesis:", general_hypothesis)

```
